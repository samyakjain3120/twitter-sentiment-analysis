```python
import re
```

```python
import nltk
nltk.download("stopwords")
```

```python
# importing nltk objects which will be in further use.
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps =  PorterStemmer()
```

```python
corpus=[]
for i in range(len(message)):
    review= re.sub('[^a-zA-Z]', ' ', col1[i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review= ' '.join(review)
    corpus.append(review)
```

# Word Embedding

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size= 1000
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
```

```python
print(onehot_repr)
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import SpatialDropout1D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Bidirectional
```

```python
sent_length= 15
```

```python
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs[6])
```

```python
dim=50
```

```python
embedding_vector_features = 50

model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(SpatialDropout1D(0.4))
model.add(Bidirectional(LSTM(200, dropout=0.2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
X = np.array(embedded_docs)
```

```python
X.shape
```

```python
y = message['label'].values
```

```python
y.shape
```

Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

# LSTM Model

```python
model.fit(X_train,y_train, epochs=7,validation_data=(X_test,y_test), batch_size=32)
```

## Test data

```python
y_pred=model.predict_classes(X_test)
y_pred
```

```python

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
